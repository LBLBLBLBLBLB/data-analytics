```python
import pandas as pd
df = pd.read_csv('survey_results_public.csv')
df.head()
```

```python
# კონკრეტული სვეტებისა და სტრიქონების გამოტანა
selected_data = df[['Age','YearsCode','DevType','LearnCode','EdLevel','LanguageHaveWorkedWith', 'LanguageWantToWorkWith',
                    'ConvertedCompYearly']]
selected_data.head()
```

```python
# ცხრილის ინდექსირება კონკრეტული სვეტის (DevType) მიმართ
selected_data.set_index('DevType')
selected_data.head()
```

```python
#ცხრილის გაფილტვრა ორ პარამეტრზე დამოკიდებული ფილტრით 
filter = (selected_data['LanguageHaveWorkedWith'].str.contains('Python')) & (selected_data['Age']=='Under 18 years old')
selected_data.loc[filter]
```

```python
# სორტირება ცხრილის ორი პარამეტრის მიხედვით
sorted_data =selected_data.sort_values(by=['YearsCode', 'LearnCode'])
sorted_data
```

```python
# მონაცემთა მანიპულაცია
# mean ფუნქციით საშუალოს პოვნა
mean_values = selected_data[['ConvertedCompYearly']].mean()
# std ფუნქციით სტანდარტული გადახრის პოვნა
std_deviation_values = selected_data[['ConvertedCompYearly']].std()

#მედიანის, მინიმალური და მაქსიმალური მნიშვნელობების პოვნა
median_values = selected_data[['ConvertedCompYearly']].median()
min_values = selected_data[['ConvertedCompYearly']].min()
max_values = selected_data[['ConvertedCompYearly']].max()

print("Mean Values:" ,mean_values)
print("\nStandard Deviation Values:", std_deviation_values)
print("\nMedian Values:", median_values)
print("\nMinimum Values:",min_values)
print("\nMaximum Values:",max_values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# NaN მნიშვნელობების მოშორება
selected_data = selected_data.dropna(subset=['LanguageHaveWorkedWith'])

# dict ში, LanguageHaveWorkedWith სვეტის გამოყენებით, პროგრამირების ენების და მათი რაოდენობის შენახვა
language_count = {}
for row in selected_data['LanguageHaveWorkedWith']:
    languages = row.split(';')
    for language in languages:
        language_count[language] = language_count.get(language, 0) + 1
        
languages = list(language_count.keys())
counts = list(language_count.values())        
original_order = np.arange(len(languages))

# bar ტიპისი დიაგრამა პროგრამირების ენებზე რომლებთანაც უმუშავიათ
plt.figure(figsize=(20, 18))  

bar_width = 1.5  
spacing = 0.5  


y_positions = original_order * (bar_width + spacing)

plt.barh(y_positions, counts, height=bar_width, align='center')

plt.yticks(y_positions, languages,fontsize=14)
plt.xlabel('Count',fontsize=15)
plt.ylabel('Programming Language',fontsize=15)
plt.title('Count of Programming Languages',fontsize=18)

plt.show()
```

```python
selected_data = selected_data.dropna(subset=['EdLevel'])

education_counts = selected_data['EdLevel'].value_counts()

# წრიული დიაგრამ განათლების დონის შესახებ
plt.figure(figsize=(12, 12))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(education_counts))))
plt.title('Distribution of Education Levels')

plt.show()
```
