```python
import pandas as pd
```

```python
# N1. DATA CLEANING
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df
```

```python
# combine magnitude and magnitude type
df['Magnitude and Magnitude Type'] = df['Magnitude'].astype(str) + " "+ df['Magnitude Type'].astype(str)
df
```

```python
# remove columns (magnitude source)
df = df.drop(columns = 'Magnitude Source') 
df
```

```python
# replace nan with 0 
columns_to_fill = ["Depth Error", "Magnitude Error", "Magnitude Seismic Stations", "Azimuthal Gap",
                   "Horizontal Distance", "Horizontal Error", "Root Mean Square"]
df[columns_to_fill] = df[columns_to_fill].fillna(0)
df
```

```python
# create new columns from date
df[["Month", "Day","Year"]] = df['Date'].str.extract('(\d+)/(\d+)/(\d+)')
df
```

```python
#N2. ADD NEW CATEGORY
```

```python
# create categories for depth and add new column for it
bin_edges = [0, 30, 60, float('inf')]
bin_labels = ["Shallow", "Intermediate", "Deep"]
df['Depth_Category'] = pd.cut(df['Depth'], bins=bin_edges, labels=bin_labels, include_lowest=True)
df
```

```python
#N3. DATA RESHAPING
```

```python
melted_df = pd.melt(df, id_vars=["Date", "Time", "Latitude", "Longitude", "Type", "Depth", "Depth Error", "Depth Seismic Stations",
                                  "Magnitude Error", "Magnitude Seismic Stations"],
                    var_name="Variable", value_name="Value")
print("Melted DataFrame:")
print(melted_df)
```

```python
pivot_table_df = df.pivot_table(index="Type", columns="Year", values="Depth", aggfunc="mean")
print("Pivot Table:")
print(pivot_table_df)
```

```python
#N4. CHARTS
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df['Status'].value_counts().plot(kind='bar', color='coral')
plt.title('Distribution of Earthquake Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])
df['Year'] = df['Date'].dt.year
avg_magnitude_by_year = df.groupby('Year')['Magnitude'].mean().reset_index()


plt.figure(figsize=(12, 6))
plt.plot(avg_magnitude_by_year['Year'], avg_magnitude_by_year['Magnitude'], marker='o', linestyle='-', color='red')
plt.title('Average Magnitudes by Year')
plt.xlabel('Year')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Horizontal Distance'], df['Horizontal Error'], color='purple', alpha=0.5)
plt.title('Scatter Plot of Horizontal Distance vs Horizontal Error')
plt.xlabel('Horizontal Distance')
plt.ylabel('Horizontal Error')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Depth_Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, explode=(0.1, 0, 0))
plt.title('Distribution of Depth Categories')
plt.show()
```
